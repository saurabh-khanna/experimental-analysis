```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
```

```python
# Dataset for Q2
df = pd.read_stata('educ430_pset1_prob2.dta')
df.describe()
```

```python
df.head()
```

```python
# 2a
smf.ols(formula = 'scyfnsh ~ C(prsch_c)', data = df).fit().summary()
```

```python
# checking with t-test as well (looks good!)
x1 = df[df['prsch_c'] == 0]
x2 = df[df['prsch_c'] == 1]
x1 = x1['scyfnsh']
x2 = x2['scyfnsh']
diff = x1.mean(skipna = True) - x2.mean(skipna = True)
print("Mean diff: ", diff)
print("P-value for t-test: ", stats.ttest_ind(x1.values, x2.values, nan_policy = 'omit').pvalue)
```
