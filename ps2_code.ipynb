```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
from linearmodels.system import SUR
from scipy.stats import ttest_1samp
```

```python
# Dataset for Q1
df = pd.read_stata('pset2.dta')
df.describe()
```

```python
df.head()
```

```python
# 1.1
# white names
df_white = df[df['race'] == 'w']
smf.probit(formula = 'call ~ yearsexp + yearsexp2 + volunteer + military + email + empholes + workinschool + honors + computerskills + specialskills + boston + female + manager + supervisor + secretary + salesrep + retailsales + offsupport + req + expreq + comreq + educreq + compreq + orgreq', data = df_white).fit(cov_type='cluster', cov_kwds={'groups': df_white['adid']}).get_margeff().summary()
```

```python
# black names
df_black = df[df['race'] == 'b']
smf.probit(formula = 'call ~ yearsexp + yearsexp2 + volunteer + military + email + empholes + workinschool + honors + computerskills + specialskills + boston + female + manager + supervisor + secretary + salesrep + retailsales + offsupport + req + expreq + comreq + educreq + compreq + orgreq', data = df_black).fit(cov_type='cluster', cov_kwds={'groups': df_black['adid']}).get_margeff().summary()
```
